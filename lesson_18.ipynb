```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

columns = [
    ('total_area', 'Total Area Distribution'),
    ('rooms', 'Rooms Distribution'),
    ('floor', 'Floor Distribution'),
    ('floors_total', 'Floors Total Distribution'),
    ('ceiling_height', 'Ceiling Height Distribution'),
    ('living_area', 'Living Area Distribution'),
    ('kitchen_area', 'Kitchen Area Distribution'),
]

fig = make_subplots(rows=len(columns),
                    cols=1,
                    shared_xaxes=False,
                    vertical_spacing=0.05,
                    subplot_titles=[title for _, title in columns])

for i, (col, title) in enumerate(columns, start=1):
    fig.add_trace(
        go.Box(x=df[col],
               orientation='h',
               name=title,
               boxmean='sd'),
               row=i,
               col=1
    )

fig.update_layout(
    height=300 * len(columns),
    showlegend=False,
    title_text='Distribution of Real Estate Features'
)
```

```python
sum_parts = df.living_area + df.kitchen_area
correlation_factor = df.total_area / sum_parts
mask = df.total_area < sum_parts

df.loc[mask, 'living_area'] *= correlation_factor
df.loc[mask, 'kitchen_area'] *= correlation_factor
```

```python
df['living_ratio'] = df.living_area / df.total_area
df['kitchen_ratio'] = df.kitchen_area / df.total_area

```
