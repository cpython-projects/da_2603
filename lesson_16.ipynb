```python
df = pd.read_csv('session_04_data.csv')
df
```

```python
df.head()
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
df.head()
```

```python
df.columns.tolist()
```

```python
columns = {
    'ID': 'id',
    'Name': 'user_name',
    'Age': 'age',
    'Salary ($)': 'user_salary',
    'Hire Date': 'hire_date',
    'Country': 'country',
    'Weight': 'weight'
}
df.rename(columns=columns, inplace=True)
df.head()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
df.age.unique()
```

```python
df.age.replace({'Thirty': 30, 'Twenty': 20}, inplace=True)
df.age.unique()
```

```python
df['age'] = pd.to_numeric(df.age, errors='coerce')
```

```python
df.age.unique()
```

```python
df['age']  = df.age.fillna(df.age.mean())
df.age.unique()
```

```python
df.hire_date.unique()
```

```python
from dateutil import parser

def date_parser(item):
  if pd.isna(item): # pd.notna(item)
    return item
  return parser.parse(item).strftime('%Y-%m-%d')

df['hire_date'] = df.hire_date.apply(date_parser)
df.hire_date.unique()
```

```python
df.hire_date = pd.to_datetime(df.hire_date, format='%Y-%m-%d')
df.hire_date.unique()
df.info()
```

```python
df.country.unique()
```

```python
df.country = df.country.replace({'U.S.A.': 'USA', 'United Kingdom': 'UK'})
df.country = df.country.str.strip().str.title()
df.country.unique()
```

```python
def convert_weigh(item):
  if pd.isna(item):
    return item

  if 'kg' in item.lower():
    return float(item.replace('kg', ''))

  if 'lbs' in item.lower():
    return float(item.replace('lbs', '')) * 0.453592

  return item


df.weight = df.weight.apply(convert_weigh)
```

```python
df.gender = df.gender.astype('category')
df.country = df.country.astype('category')
df.info()
```

```python
df.user_salary.fillna(df.user_salary.median(), inplace=True)
df.info()
```
