```python
def add(a, b):
  res = a + b
  return res


x = add(4, 5)
print(x)

x = add(40, 50)
print(x)

x = add('4', '5')
print(x)
```

```python
def salary_info(data):
  min_salary = min(data)
  max_salary = max(data)
  avg_salary = sum(data) / len(data)
  return min_salary, max_salary, avg_salary


data = [200, 300, 432, 123, 345, 1256]
min_salary, max_salary, avg_salary = salary_info(data)

print(f"Minimum salary: {min_salary}")
print(f"Maximum salary: {max_salary}")
print(f"Average salary: {avg_salary}")
```

```python
def arithmetic_results(a, b):
  return a + b, a - b, a * b, a / b


res = arithmetic_results(10, 5)
x1, x2, x3, x4 = res
print(res)
print(x1, x2, x3, x4)
```

```python
def calc(a, b, operation):
  if operation == '+':
    return a + b
  if operation == '-':
    return a - b
  if operation == '*':
    return a * b
  if operation == '/':
    return a / b

  return 'Unknown operation'

res = calc(10, 5, '-')
print(res)
```

```python
prices = [100, 230, 3440, 3434, 3435, 2323]
rate = 1.2
res = []
for item in prices:
  res.append(item * rate)
print(res)
```

```python
prices = [100, 230, 3440, 3434, 3435, 2323]
rate = 1.2
res = [item * rate for item in prices]
print(res)
```

```python
def price_convertor(prices, rate):
  res = [item * rate for item in prices]
  return res

prices = [100, 230, 3440, 3434, 3435, 2323]
prices_new = price_convertor(prices, 1.2)
print(prices_new)
```

```python
def filter_positive(data):
  res = [item for item in data if item > 0]
  return res

data_new = filter_positive([100, -230, 3440, -3434, 3435, -2323])
print(data_new)
```

```python
def expenses(file_name):
  file = open(file_name)
  total = 0
  for line in file:
    amount = line.split()[-2].replace('$', '')
    total += float(amount)
  file.close()
  return total

res = expenses('09_test.txt')
print(f'Total: {res:.2f}$')
```

```python
from collections import defaultdict
def expenses_by_category(file_name):
  file = open(file_name)
  categories = defaultdict(float)
  for line in file:
    *_, amount, category = line.split()
    amount = amount.replace('$', '')
    categories[category] += float(amount)
  file.close()
  return categories

res = expenses_by_category('09_test.txt')
for category, amount in res.items():
  print(f'{category}: {amount:.2f}$')
```
