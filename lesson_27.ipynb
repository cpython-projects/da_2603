```python
base_url = 'https://coinmarketcap.com/'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0 Safari/537.36',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9',
    'Accept-Language': 'en-US,en;q=0.5',
    'Referer': 'https://www.google.com/',
}


responce = requests.get(base_url, headers=headers)
```

```python
responce.status_code
```

```python
responce.text
```

```python
soup = BeautifulSoup(responce.text, 'html.parser')
print(soup.prettify())
```

```python
# <table class="sc-db1da501-3 hvtTnJ cmc-table cmc-table-in-touch-device">

table = soup.find('table', class_='cmc-table')
```

```python
print(table.prettify())
```

```python
columns_names = table.thead.tr.find_all('th')
for item in columns_names:
  print(item.text)
```

```python
tbody = table.tbody.find_all('tr')


data = []

for item in tbody:
  try:
    cells = item.find_all('td')
    if not cells or len(cells) < 4:
      continue
    rank = cells[1].text.strip()
    name = cells[2].select_one('p.coin-item-name').text.strip()
    symbol = cells[2].select_one('p.coin-item-symbol').text.strip()
    price = cells[3].text.strip()

    data.append({
        'rank': rank,
        'name': name,
        'symbol': symbol,
        'price': price
        })
  except:
    continue
```

```python
import pandas as pd

df = pd.DataFrame(data)

df
```

```python
# https://coinmarketcap.com/?page=3
import time

i = 1
res = []
while True:

  url = f'https://coinmarketcap.com/?page={i}'
  soup = BeautifulSoup(requests.get(url, headers=headers).text, 'html.parser')
  table = soup.find('table', class_='cmc-table')
  if not table:
    break
  tbody = table.tbody.find_all('tr')

  if not tbody:
    break

  tbody = table.tbody.find_all('tr')
  data = []

  for item in tbody:
    try:
      cells = item.find_all('td')
      if not cells or len(cells) < 4:
        continue
      rank = cells[1].text.strip()
      name = cells[2].select_one('p.coin-item-name').text.strip()
      symbol = cells[2].select_one('p.coin-item-symbol').text.strip()
      price = cells[3].text.strip()

      data.append({
          'rank': rank,
          'name': name,
          'symbol': symbol,
          'price': price
          })
    except:
      continue

  df = pd.DataFrame(data)

  res.append(df)
  i += 1
  time.sleep(3)
```
