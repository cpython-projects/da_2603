```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv', delimiter=',', encoding='utf-8')
```

```python
df
```

```python
df.info()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df.Embarked.unique()
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.drop(columns=['Cabin', 'Name', 'Ticket'], inplace=True)
```

```python
df.info()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np
```

```python
df.head(3)
```

```python
fig = px.histogram(df, x='Age', nbins=100, title='Age Distribution')
fig.show()
```

```python
sns.histplot(df['Age'], bins=20, kde=True, color='skyblue')
plt.xlabel('Age')
plt.ylabel('count')
plt.title('Age Distribution')
plt.show()
```

```python
fig = px.box(df, x='Pclass', y='Age', title='Age Distribution by Passenger Class')
fig.show()
```

```python
sns.boxplot(x='Pclass', y='Age', data=df)
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.title('Age Distribution by Passenger Class')
plt.show()
```

```python
fig = px.violin(df, x='Pclass', y='Age', box=True, title='Age Distribution by Passenger Class')
fig.show()
```

```python
sns.violinplot(x='Pclass', y='Age', data=df)
plt.xlabel('Passenger Class')
plt.ylabel('Age')
```

```python
sns.countplot(x='Sex', hue='Pclass', data=df)
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.title('Passenger Class Distribution')
plt.show()
```

```python
fig = px.histogram(df, x='Sex', color='Pclass', barmode='group', title='Fare Distribution')
fig.show()
```

```python
fig = px.scatter(df, x='Age', y='Fare', color='Survived', title='Age vs. Fare')
fig.show()
```

```python
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs. Fare')
plt.show()
```
